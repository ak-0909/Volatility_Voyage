```python
import pandas as pd
```

```python
df = pd.read_csv('data-27.csv')
df.head()
```

PART 1

```python
print("Shape of the dataset:", df.shape)

print("\nColumns in dataset:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nSummary Statistics:")
print(df.describe())

print("\nMissing Values in each column:")
print(df.isnull().sum())
```

```python
df['SMA'] = df['close'].rolling(window=20).mean()

df['EMA_20'] = df['close'].ewm(span=20, adjust=False).mean()
df['EMA_50'] = df['close'].ewm(span=50, adjust=False).mean()

df[['close', 'SMA', 'EMA_20', 'EMA_50']].head(30)
```

PART 2

```python
df['date'] = pd.to_datetime(df['datetime'])

df.set_index('date', inplace=True)

df.head()
```

```python
volatility = df['close'].resample('6M').std()

print("🔹 6-Month Standard Deviations:\n")
print(volatility)
```

```python
most_volatile_period = volatility.idxmax()
highest_std = volatility.max()

print("\nMost Volatile 6-Month Period:")
print(f"Start of Period: {most_volatile_period.strftime('%Y-%m-%d')}")
print(f"Standard Deviation: {highest_std:.2f}")
```
